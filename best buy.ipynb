```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
driver=webdriver.Chrome()
driver.get("https://www.bestbuy.com/site/promo/laptop-and-computer-deals")
driver.maximize_window()
wait=WebDriverWait(driver, 25)
```

```python
driver.find_element(by=By.XPATH,value='/html/body/div[2]/div/div/div/div[1]/div[2]/a[2]').click()
```

```python
products=driver.find_elements(by=By.CSS_SELECTOR,value='ol.sku-item-list li.sku-item div div div div div div.column-left a')
```

```python
product_links=[product.get_attribute('href') for product in products]
```

```python
data=[]
for i,prod_link in enumerate(product_links):
    driver.get(prod_link)
    try:
        name=driver.find_element(by=By.CSS_SELECTOR,value='div.sku-title h1').text
    except:
        name=driver.find_element(by=By.CSS_SELECTOR,value="h1.h4").text
        
    try:
        price=wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'[data-testid="customer-price"] span'))).text
    except:
        price=wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'[data-testid="large-customer-price"]'))).text

    # img_button=driver.find_element(by=By.CSS_SELECTOR,value="li.image-more-thumbnail.mb-100.mr-400")
    # driver.execute_script("arguments[0].click();", img_button)
    # driver.execute_script("arguments[0].scrollIntoView(true);", img_button)
    # img_button.click()
    img_links=[]
    try:
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'li.image-more-thumbnail.mb-100.mr-400'))).click()
        some=driver.find_element(by=By.ID,value='product-images-content-panel')
        image_elements=some.find_elements(by=By.CSS_SELECTOR,value='div div div div div div ol li div button')

        for image_element in image_elements:
            try:
                time.sleep(1)
                image_element.click()
            except:
                # driver.find_element(by=By.XPATH,value='//*[@id="product-images-content-panel"]/div/div[1]/div/div/div[3]/div/button').click()
                # driver.execute_script("arguments[0].scrollIntoView(true);", img_element)
                # img_element.click()
                pass
            time.sleep(1)
            wait2=WebDriverWait(some, 25)
            img=wait2.until(EC.presence_of_element_located((By.CSS_SELECTOR,'div div div [data-testid="hammer-div"] button img')))
            images=some.find_element(by=By.CSS_SELECTOR,value='div div div [data-testid="hammer-div"] button img')
            img_links.append(images.get_attribute('src'))
        
        driver.find_element(by=By.XPATH,value='//*[@id="shop-media-gallery-portal"]/div/div/div[1]/div/div/div/button').click()
    except:
        driver.find_element(by=By.CSS_SELECTOR,value='div.list-none.relative.h-800.w-800.m-50 button').click()
        img_buttons=driver.find_elements(by=By.CSS_SELECTOR,value='div.pl-flex-carousel-slider ul li div button')
        for i,img_button in enumerate(img_buttons):
            img_buttons[i].click()
            img=driver.find_elements(by=By.CSS_SELECTOR,value='div.flex.h-full img')
            if img[i].get_attribute("src").startswith("https://pisces.bbystatic.com/image2/BestBuy_US/images/products/"):
                print(img[i].get_attribute("src"))
            else:
                break
        driver.find_element(by=By.XPATH,value='/html/body/div[4]/div[4]/div[1]/div[2]/div[3]/div[1]/div/div/div/div/button').click()
    time.sleep(2)
    try:
        driver.find_element(by=By.CSS_SELECTOR,value='div.shop-specifications div div div div.mb-600 button').click()
        specs=driver.find_elements(by=By.CSS_SELECTOR,value='div.c-overlay-backdrop div div div ul li div div.w-full')
        details={}
        for spec in range(0, len(specs), 2):
            details[specs[spec].text]=specs[spec+1].text
            
        driver.find_element(by=By.XPATH,value='//*[@id="pdp-drawer-overlay-backdrop"]/div/div/div[1]/button').click()
        
    except:
        driver.find_element(by=By.CSS_SELECTOR,value="div.v-border-bottom button").click()
        specs=driver.find_elements(by=By.CSS_SELECTOR,value='[aria-label="Specifications"] div div [data-testid="brix-sheet-content"] ul li div div')
        details={}
        for spec in range(0, len(specs), 2):
            details[specs[spec].text]=specs[spec+1].text

        driver.find_element(by=By.XPATH,value='/html/body/div[4]/div[4]/div[1]/div[4]/div/div/div[2]/div/div[1]/div/button').click()

    # time.sleep(2)
    
    # try: 
    element = driver.find_element(By.CSS_SELECTOR, "div.shop-overview-accordion div button")
    driver.execute_script("arguments[0].click();", element)
    #     try:
    feats = driver.find_elements(by=By.CSS_SELECTOR,value='[data-testid="pdp-utils-product-info"] > ul:nth-of-type(2) li')
    features=[]
    for feat in feats:
        # features[feat.text.split('\n')[0]]=feat.text.split('\n')[1]
        features.append(feat.text)
    #     except:
    #         pass

    # driver.find_element(by=By.XPATH,value='//*[@id="pdp-drawer-overlay-backdrop"]/div/div/div[1]/button').click()
    # except:
    #     elem=wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,"div.v-border-bottom button")))
    #     elem[1].click()
    #     feats=driver.find_elements(by=By.CSS_SELECTOR,value='[data-testid="brix-sheet-content"] ul.pl-300 li')
    #     for feat in feats:
    #         features[feat.text.split('\n')[0]]=feat.text.split('\n')[1]
                
    #     driver.find_element(by=By.XPATH,value='//*[@id="pdp-drawer-overlay-backdrop"]/div/div/div[1]/button').click()


    print(i)
    
    data.append({'name':name,'price':price,'img':",".join(img_links),**details,'features':features})
```
